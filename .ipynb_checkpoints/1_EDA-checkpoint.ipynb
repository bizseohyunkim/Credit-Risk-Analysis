```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.makedirs('images', exist_ok=True)

df = pd.read_csv(r'C:\Users\playdata2\Desktop\Credit-Risk-Analysis\data\creditcard.csv')
print(f"데이터 크기: {df.shape}")
df.head()
```

```python
print("=== 기본 정보 ===")
print(df.info())
print("\n=== 결측치 ===")
print(df.isnull().sum())
print("\n=== 사기 비율 ===")
print(df['Class'].value_counts())
print(df['Class'].value_counts(normalize=True) * 100)
```

```python
plt.figure(figsize=(6,4))
counts = df['Class'].value_counts()
plt.bar(['정상(0)', '사기(1)'], counts.values, color=['steelblue', 'crimson'])
plt.title('정상 vs 사기 거래 수')
plt.ylabel('거래 수')
for i, v in enumerate(counts.values):
    plt.text(i, v + 100, f'{v:,}건\n({v/len(df)*100:.2f}%)', ha='center')
plt.tight_layout()
plt.savefig('images/class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df[df['Class']==0]['Amount'].hist(ax=axes[0], bins=50, color='steelblue', edgecolor='white')
axes[0].set_title('정상 거래 금액 분포')
axes[0].set_xlabel('Amount (USD)')

df[df['Class']==1]['Amount'].hist(ax=axes[1], bins=50, color='crimson', edgecolor='white')
axes[1].set_title('사기 거래 금액 분포')
axes[1].set_xlabel('Amount (USD)')

plt.tight_layout()
plt.savefig('images/amount_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

df[df['Class']==0]['Time'].hist(ax=axes[0], bins=50, color='steelblue', edgecolor='white')
axes[0].set_title('정상 거래 시간 분포')
axes[0].set_xlabel('Time (초)')

df[df['Class']==1]['Time'].hist(ax=axes[1], bins=50, color='crimson', edgecolor='white')
axes[1].set_title('사기 거래 시간 분포')
axes[1].set_xlabel('Time (초)')

plt.tight_layout()
plt.savefig('images/time_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr = df.corr()['Class'].drop('Class').sort_values()
corr.plot(kind='barh', color=['crimson' if x < 0 else 'steelblue' for x in corr])
plt.title('각 변수와 사기(Class)의 상관관계')
plt.xlabel('상관계수')
plt.tight_layout()
plt.savefig('images/correlation.png', dpi=150, bbox_inches='tight')
plt.show()
```
